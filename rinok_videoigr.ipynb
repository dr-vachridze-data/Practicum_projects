**Пропущенные значения**

- Максимальное значение пропущенных данных в столбце 'User Score' - 54%, на втором месте столбец 'Critic Score' - 51%, на третем месте столбец 'Rating' - 40%

- Пропуски в колонке 'Name' не критичны, поэтому их можно игнорировать или удалить, т.к их кол-во менее 1%.
- Пропуски в колонке 'Genre' не критичны, поэтому их можно игнорировать или удалить, т.к их кол-во менее 1%.
- Пропуски в колонке 'EU sales' и 'JP sales' можно заменить на среднее (в зависимости от названия платформы и года выхода игры)
- Пропуски в колонке 'Critic Score' не критичны, можно игнорировать.
- Пропуски в колонке 'User Score' не критичны, можно игнорировать.
- Пропуски в колонке 'Rating' не критичны, можно игнорировать.

- Обработайте пропущенные значения. Для каждого случая вы можете выбрать оптимальный, на ваш взгляд, вариант: заменить на определённое значение, оставить как есть или удалить.
- Если вы решите заменить пропуски на значение-индикатор, то убедитесь, что предложенное значение не может быть использовано в данных.
- Если вы нашли пропуски в данных с количеством проданных копий игры в том или ином регионе, их можно заменить на среднее значение в зависимости от названия платформы и года выхода игры.

```python
#Убираем пропуски в 'EU sales' - среднее значение в зависимости от названия платформы и года выхода игры
def group_eu_sales(row):
    group = df[(df['platform'] == row['platform']) & 
               (df['year_of_release'] == row['year_of_release'])]
    return group['eu_sales'].mean()
df['eu_sales'].fillna('group_eu_sales', inplace=True)
```

```python
#Убираем пропуски в 'JP sales' - среднее значение в зависимости от названия платформы и года выхода игры
def group_jp_sales(row):
    group = df[(df['platform'] == row['platform']) & 
               (df['year_of_release'] == row['year_of_release'])]
    return group['jp_sales'].mean()
df['jp_sales'].fillna('group_jp_sales', inplace=True)
```

```python
#Удаляем пропуски в 'Name' и 'Genre'
df = df.dropna(subset=['name', 'genre'])
```

## 3. Фильтрация данных

- Данные за период с 2000 по 2013 год включительно. 

```python
df_actual=df_no_duplicates.loc[(df_no_duplicates['year_of_release'] >=2000) & (df_no_duplicates['year_of_release'] <=2013)]
```

```python
df_actual.info()
```

```python
df_actual=df_actual.copy()
```

## 4. Категоризация данных
    
- Проводим категоризацию данных по оценкам пользователей


```python
def user_categories(cat):
    if cat <3:
        return 'низкая оценка'
    elif cat <8:
        return 'средняя оценка'
    elif cat <=10:
        return 'высокая оценка'
df_actual['user_category']=df_actual['user_score'].apply(user_categories)
```

- Проводим категоризацию данных по оценкам критиков.

```python
def critic_categories(cat):
    if cat <30:
        return 'низкая оценка'
    elif cat <80:
        return 'средняя оценка'
    elif cat <=100:
        return 'высокая оценка'
df_actual['critic_category']=df_actual['critic_score'].apply(critic_categories)
```

```python
df_actual.info()
```

```python
#Кол-вр игры по категориям среди пользователей
grouped_df_user = df_actual.groupby('user_category')['name'].count()
print(grouped_df_user)
```

```python
#Кол-вр игры по категориям среди критиков
grouped_df_critic = df_actual.groupby('critic_category')['name'].count()
print(grouped_df_critic)
```

- Топ-7 платформ по количеству игр, выпущенных за весь актуальный период.

```python
top_7 = df_actual.groupby('platform').size().sort_values(ascending=False)
print(top_7.head(7))
```

За период с 2000 по 2013 год в Топ-7 платформ по количеству выпущенных игр вошли такие платформы как: PS2, DS, Wii, PSP, X360, PS3, GBA.

Лидирующие позиции с небольшим отрывом заняли платформы PS2 и DS, заняв большую долю выпущенных игр.
